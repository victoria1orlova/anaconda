```python
import matplotlib.pyplot as plt
import numpy as np

# Генерація значень x та y
x = np.linspace(-1.5, 1.5, 400)
y = np.linspace(-2, 3, 400)
X, Y = np.meshgrid(x, y)

# Рівняння функції
F = X**2 + (Y - np.sqrt(np.abs(X)))**2 - 1

# Побудова контурного графіку
plt.figure(figsize=(6, 6))  # Налаштування розміру графіку
plt.contour(X, Y, F, levels=[0], colors='r')
plt.title('$x^2 + (y - sqrt(|x|))^2 = 1$')  # Заголовок графіку
plt.xlabel('x')  # Підпис осі x
plt.ylabel('y')  # Підпис осі y
plt.grid(True)
plt.axis('equal')  # Збереження пропорційності
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Задання діапазону для параметра t
t = np.linspace(0, 2 * np.pi, 1000)  # Від 0 до 2π

# Обчислення x(t) та y(t) для кожного значення t
x = 4 * np.sin(t) - np.sin(3 * t) - 0.2 * np.cos(10.5 * t)
y = 4 * np.cos(t) - 2 * np.cos(2 * t) - np.cos(3 * t) + 0.15 * np.sin(8 * t)

# Побудова графіку
plt.figure(figsize=(8, 8))  # Налаштування розміру графіку
plt.plot(x, y, label='Траєкторія (x(t), y(t))')
plt.title('Параметрична траєкторія (x(t), y(t))')
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Генерація значень параметра t
t = np.linspace(0, 10 * np.pi, 1000)  # від 0 до 10π

# Рівняння для x(t) та y(t)
x = 4 * np.sin(t) - np.sin(3 * t) - 0.2 * np.cos(10.5 * t)
y = 4 * np.cos(t) - 2 * np.cos(2 * t) - np.cos(3 * t) + 0.15 * np.sin(8 * t)

# Побудова графіку траєкторії
plt.figure(figsize=(6, 6))  # Налаштування розміру графіку
plt.plot(x, y, color='purple', label='Криві x(t) та y(t)')
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()
plt.show()
```
