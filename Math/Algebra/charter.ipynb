```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.linspace(-10, 10, 100)  # Create an array of 100 points from -10 to 10

# Calculate y values (x^2)
y = x**2

# Create the plot
plt.plot(x, y)
plt.title("Chart of x^2")
plt.xlabel("x")
plt.ylabel("y = x^2")

# Show the plot
plt.grid(True)
plt.show()
```
