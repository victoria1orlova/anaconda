```python
import matplotlib.pyplot as plt
import numpy as np

# Генеруємо значення x
x = np.linspace(0, 10, 50)  # Створюємо масив із 50 точок від 0 до 10

y = (5-x)**2

# Створюємо графік для y = x^2
plt.plot(x, y, label='y')

# Додаємо підписи до графіку
plt.title("Графік параболи y = x^2")
plt.xlabel("x")
plt.ylabel("y")

# Показуємо легенду графіка
plt.legend()

# Відображаємо графік
plt.grid(True)
plt.show()
```
