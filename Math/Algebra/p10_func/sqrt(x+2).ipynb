```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.linspace(-2, 10, 100)  # Create an array of 100 points from -10 to 10

y = np.sqrt(x+2)

# Create the plot
plt.plot(x, y)
plt.title("Chart")
plt.xlabel("x")
plt.ylabel("y")

# Show the plot
plt.grid(True)
plt.show()
```
