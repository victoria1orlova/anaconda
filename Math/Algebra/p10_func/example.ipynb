```python
import matplotlib.pyplot as plt
import numpy as np

# Генеруємо значення x
x = np.linspace(-2, 6, 40)  # Створюємо масив із 100 точок від -10 до 14

# Розраховуємо значення y (f(x) = x^2 - 4x - 5)
y = x**2 - 4*x - 5

# Створюємо графік
plt.plot(x, y, label='f(x) = x^2 - 4x - 5')

# Додаємо підписи до графіку
plt.title("Графік функції f(x) = x^2 - 4x - 5")
plt.xlabel("x")
plt.ylabel("f(x)")

# Показуємо легенду графіка
plt.legend()

# Відображаємо графік
plt.grid(True)
plt.show()
```
