```python
import matplotlib.pyplot as plt
import numpy as np

# Визначте кількість вершин (у вас шестикутник, тому 6 вершин)
num_vertices = 6

# Визначте центр багатокутника
center = [10, 10]  # Ваші координати центру

# Визначте радіус (відстань від центру до грані)
radius = 25.0  # Ваша відстань до грані

# Обчисліть координати вершин багатокутника
angle = 2 * np.pi / num_vertices
vertices = []
for i in range(num_vertices):
    x = center[0] + radius * np.cos(i * angle)
    y = center[1] + radius * np.sin(i * angle)
    vertices.append([x, y])

# Побудова багатокутника
polygon = plt.Polygon(vertices, closed=True, edgecolor='b', facecolor='none')

# Створення графіку та додавання багатокутника
fig, ax = plt.subplots()
ax.add_patch(polygon)

# Побудова кола
circle = plt.Circle(center, radius, edgecolor='r', facecolor='none')

# Додавання кола до графіку
ax.add_patch(circle)

# Побудова кола
'''
circumscribed circle
inscribed circle
'''
# Обчисліть радіус вписаного кола
inradius = radius * np.cos(np.pi / num_vertices)
circle_inscribed = plt.Circle(center, inradius, edgecolor='g', facecolor='none')

# Додавання кола до графіку
ax.add_patch(circle_inscribed)

# Налаштування графіку
ax.set_aspect('equal')
ax.set_xlim(center[0] - radius, center[0] + radius)
ax.set_ylim(center[1] - radius, center[1] + radius)

# Виведення графіку на екран
plt.show()
```

```python
# Обчислення довжини граней многоугольника
side_length = 2 * radius * np.sin(np.pi / num_vertices)
print(f"Довжина грані багатокутника: {side_length:.2f}")

# Обчислення радіуса вписаного кола
inradius = inradius * np.cos(np.pi / num_vertices)
print(f"Радіус вписаного кола: {inradius:.2f}")

# Обчислення радіуса описаного кола
circumradius = radius
print(f"Радіус описаного кола: {circumradius:.2f}")
```
