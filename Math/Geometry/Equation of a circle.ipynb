```python
import matplotlib.pyplot as plt
import numpy as np

# Створення точок для побудови кола
theta = np.linspace(0, 2*np.pi, 100)  # 100 точок для наближеного кола
r = 1  # Радіус кола

x = r * np.cos(theta)
y = r * np.sin(theta)

# Побудова кола
plt.figure(figsize=(5, 5))  # Розмір графіка
plt.plot(x, y)
plt.gca().set_aspect('equal', adjustable='box')  # Збереження пропорційності
plt.title('Коло з рівнянням $x^2 + y^2 = 4$')
plt.grid(True)

# Відображення графіку
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Задане рівняння кола
radius = 1
center_x = 1
center_y = 2

# Генерація координат для кола
theta = np.linspace(0, 2 * np.pi, 100)  # 100 точок для наближеності кола
x = center_x + radius * np.cos(theta)
y = center_y + radius * np.sin(theta)

# Побудова кола
plt.figure(figsize=(5, 5))  # Налаштування розміру графіку
plt.plot(x, y, label='Коло: $x^2 + (y + 6)^2 = 4^2$')
plt.title('Побудова кола на координатній площині')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axis('equal')  # Збереження пропорційності
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Задане рівняння кола
radius = 3
center_x = 0
center_y = 0

# Генерація координат для кола
theta = np.linspace(0, 2 * np.pi, 100)  # 100 точок для наближеності кола
x = center_x + radius * np.cos(theta)
y = center_y + radius * np.sin(theta)

# Побудова кола
plt.figure(figsize=(6, 6))  # Налаштування розміру графіку
plt.plot(x, y, label='Коло: $(x-0)^2 + (y-0)^2 = 3^2$')
plt.title('Побудова кола на координатній площині')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axis('equal')  # Збереження пропорційності
plt.legend()
plt.show()
```
