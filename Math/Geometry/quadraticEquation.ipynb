```python
import cmath

def solve_quadratic(a, b, c):
    # Обчислюємо дискримінант
    D = b**2 - 4*a*c

    # Обчислюємо корені уравнення, включаючи можливі комплексні корені
    x1 = (-b + cmath.sqrt(D)) / (2*a)
    x2 = (-b - cmath.sqrt(D)) / (2*a)

    return x1, x2, D

# Приклад використання:
a = 2
b = 18
c = -72

x1, x2, discriminant = solve_quadratic(a, b, c)
print(f"Корни квадратного уравнения: x1 = {x1}, x2 = {x2}")
print(f"Дискримінант: {discriminant}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_function(a, b, c, x):
    return a*x**2 + b*x + c

# График уравнения
margin = 2  # Запас пространства вокруг корней на графике
x_range = np.linspace(min(x1.real, x2.real) - margin, max(x1.real, x2.real) + margin, 400)
y_range = quadratic_function(a, b, c, x_range)

signB = " + " if b > 0 else " - "
signC = " + " if c > 0 else " - "
plt.plot(x_range, y_range, label=f'{a}x² {signB} {abs(b)}x {signC} {abs(c)}')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("График квадратного уравнения")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.xlim([x_range[0], x_range[-1]])
plt.ylim([min(y_range), max(y_range)])
plt.legend()
plt.show()

print(f"Корни квадратного уравнения: x1 = {x1}, x2 = {x2}")
```
