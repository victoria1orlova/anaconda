```python
import matplotlib.pyplot as plt
import numpy as np

# Задане рівняння кола
radius = np.sqrt(10)
center_x = -6
center_y = 1

# Координати точки для перевірки
x_point = -3
y_point = 0

# Обчислення відстані від центру кола до точки
distance = np.sqrt((x_point - center_x)**2 + (y_point - center_y)**2)

# Перевірка, чи точка належить колу
if distance <= radius:
    print(f'Точка ({x_point}, {y_point}) належить колу.')
else:
    print(f'Точка ({x_point}, {y_point}) не належить колу.')

# Генерація координат для кола
theta = np.linspace(0, 2 * np.pi, 100)  # 100 точок для наближеності кола
x = center_x + radius * np.cos(theta)
y = center_y + radius * np.sin(theta)

# Побудова кола
plt.figure(figsize=(6, 6))  # Налаштування розміру графіку
plt.plot(x, y, label='Коло: $(x + 6)^2 + (y - 1)^2 = 10$')
plt.plot(x_point, y_point, 'ro', label='Точка (0, 1)')  # Відмічаємо точку на графіку
plt.title('Побудова кола на координатній площині')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axis('equal')  # Збереження пропорційності
plt.legend()
plt.show()
```
