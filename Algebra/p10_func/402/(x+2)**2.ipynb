```python
import matplotlib.pyplot as plt
import numpy as np

# Генeруємо область визначення x
x = np.linspace(-6, 2, 50)  # Створюємо масив із 50 точок від -2 до 2

y = (x+2)**2

# Створюємо графік для y = x^2
plt.plot(x, y, label='y')

# Додаємо підписи до графіку
plt.title("Графік параболи y")
plt.xlabel("x")
plt.ylabel("y")

# Показуємо легенду графіка
plt.legend()

# Відображаємо графік
plt.grid(True)
plt.show()
```
