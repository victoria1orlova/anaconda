```python
import matplotlib.pyplot as plt
import numpy as np

# Генеруємо значення x
x = np.linspace(-1, 7, 50)  # Створюємо масив із 50 точок від -2 до 2

y = (3-x)**2 - 1

# Створюємо графік для y = x^2
plt.plot(x, y, label='y')

# Додаємо підписи до графіку
plt.title("Графік параболи y = x^2")
plt.xlabel("x")
plt.ylabel("y")

# Показуємо легенду графіка
plt.legend()

# Відображаємо графік
plt.grid(True)
plt.show()
```
